from sklearn.cluster import KMeans: Эта строка импортирует класс KMeans из библиотеки Scikit-Learn, который представляет собой реализацию алгоритма K-средних.
from sklearn.datasets import make_blobs: Здесь импортируется функция make_blobs из Scikit-Learn, которая используется для генерации искусственных данных, предназначенных для кластеризации.
from yellowbrick.cluster import KElbowVisualizer: Эта строка импортирует класс KElbowVisualizer из библиотеки Yellowbrick. Yellowbrick - это библиотека для визуализации данных и моделей машинного обучения.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
```

X, y = make_blobs(n_samples=1000, n_features=12, centers=8, shuffle=True, random_state=42): Эта строка кода создает искусственные данные с помощью функции make_blobs. Параметры функции:
n_samples=1000: Количество образцов данных.
n_features=12: Количество признаков (переменных) в каждом образце данных.
centers=8: Количество кластеров, которые должны быть созданы.
shuffle=True: Перемешивание данных, чтобы они были случайно распределены.
random_state=42: Устанавливает начальное значение для генератора случайных чисел, что делает эксперимент воспроизводимым.

```python
X, y = make_blobs(n_samples =1000 , n_features =12, centers=8,
shuffle=True , random_state =42)
```

model = KMeans(): Создается объект model класса KMeans без указания количества кластеров. Этот объект будет использоваться для оценки оптимального числа кластеров.

```python
model = KMeans ()
```

visualizer = KElbowVisualizer(model, k=(4, 12)): Создается объект visualizer класса KElbowVisualizer, который будет использоваться для визуализации метрики "локтя" для выбора оптимального числа кластеров. Параметр k задает диапазон числа кластеров, который будет рассмотрен (от 4 до 12).
visualizer.fit(X): Обучение модели K-средних на данных X.
visualizer.show(): Визуализация метрики "локтя". Эта визуализация позволяет выбрать оптимальное количество кластеров на основе изменения инерции (суммы квадратов расстояний от точек до центроидов) по мере увеличения числа кластеров. Пользователь может выбрать количество кластеров, где инерция начинает изменяться существенно (точка "локтя"), что считается оптимальным числом кластеров.

```python
visualizer = KElbowVisualizer(model , k=(4 ,12))
visualizer.fit(X); visualizer.show()
```

Далее, код создает две дополнительные визуализации с разными метрик:
visualizer = KElbowVisualizer(model, k=(4, 12), metric='silhouette'): Эта строка создает визуализацию с использованием метрики силуэта для выбора оптимального числа кластеров.
visualizer = KElbowVisualizer(model, k=(4, 12), metric='calinski_harabasz'): Эта строка создает визуализацию с использованием метрики Калински-Харабаса для выбора оптимального числа кластеров.
После создания дополнительных визуализаций, они также обучаются на данных и отображаются с помощью visualizer.show() для каждой из них.

```python
visualizer = KElbowVisualizer(model , k=(4,12),metric='silhouette')
visualizer.fit(X); visualizer.show()
```

```python
visualizer = KElbowVisualizer(model , k=(4,12),metric='calinski_harabasz')
visualizer.fit(X); visualizer.show()
```
