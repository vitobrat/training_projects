from sklearn.datasets import make_blobs: Эта строка импортирует функцию make_blobs из библиотеки Scikit-Learn. make_blobs используется для генерации искусственных данных для задач кластеризации.
from sklearn.cluster import DBSCAN: Здесь импортируется класс DBSCAN из Scikit-Learn. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) - это алгоритм кластеризации, который основан на плотности точек данных.
import numpy as np: Эта строка импортирует библиотеку NumPy и создает псевдоним "np" для нее. NumPy используется для работы с массивами и вычислениями в Python.(check)


```python
# Import of necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np
```

X, y = make_blobs(n_samples=300, centers=[[0, 0], [0, 5], [5, 5]], cluster_std=[1.0, 0.5, 1.0]): Здесь генерируются искусственные данные с помощью функции make_blobs. Параметры функции:
n_samples=300: Количество сгенерированных образцов данных.
centers=[[0, 0], [0, 5], [5, 5]]: Координаты центров кластеров, в данном случае, создаются три кластера с указанными центрами.
cluster_std=[1.0, 0.5, 1.0]: Стандартное отклонение для каждого кластера, задающее разброс данных внутри каждого кластера.

```python
# Generate data
X, y = make_blobs(n_samples =300,
centers =[[0, 0], [0, 5], [5, 5]],
cluster_std =[1.0 , 0.5, 1.2])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

dbscan = DBSCAN(eps=0.50, min_samples=5): Здесь создается объект dbscan класса DBSCAN. Параметры алгоритма:
eps=0.50: Максимальное расстояние между двумя точками, при котором они считаются в одном кластере.
min_samples=5: Минимальное количество точек, необходимое для формирования кластера.
labels = dbscan.fit_predict(X): Этот код выполняет кластеризацию данных X с помощью алгоритма DBSCAN и присваивает каждой точке метку кластера. Метки кластеров сохраняются в переменной labels.

```python
# Compute DBSCAN
dbscan = DBSCAN(eps =0.50, min_samples =5)
labels = dbscan.fit_predict(X)
labels
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

no_noise = np.sum(np.array(labels) == -1, axis=0): Здесь используется библиотека NumPy для подсчета количества точек, которые были отнесены к шуму (несгруппированным точкам). Все точки, которые не входят ни в один кластер, имеют метку -1, и этот код находит количество таких точек и сохраняет результат в переменной no_noise.

```python
# Definition noise points
no_noise = np.sum(np.array(labels) == -1)
no_noise
```

Пример кластеризации данных, которые расположены в виде сложной формы

```python
from sklearn.datasets import make_moons
# Generate data
n_samples = 200
X, _ = make_moons(n_samples=n_samples ,
noise =0.05 ,
random_state =0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=_, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

Пример алгоритма К среднего для данной выборки

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters =2)
kmeans.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

Будем использовать поиск по сетке — метод, используемый в машинном обучении для поиска наилучшей комбинации гиперпараметров модели.
Здесь X — множество объектов, labels — метки кластеров. По умолчанию для оценки используется евклидово расстояние 𝐿2. Можно использовать
𝐿1, 𝐿∞ или косинускую меру.

```python
from sklearn.metrics import silhouette_score
silhouette = silhouette_score(X, labels)
silhouette
```
