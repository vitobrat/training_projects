```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0): Эта строка кода создает искусственные данные, которые представляют собой три кластера. Параметры функции make_blobs:
n_samples=300: Это количество образцов данных, которые будут сгенерированы.
centers=3: Задает количество центров (кластеров), которые будут созданы.
cluster_std=0.60: Этот параметр задает стандартное отклонение (разброс) данных внутри каждого кластера.
random_state=0: Устанавливает начальное значение для генератора случайных чисел, что делает эксперимент воспроизводимым.

```python
# Generate random data with 3 clusters
X, y_true = make_blobs(n_samples =300,
centers=3,
cluster_std =0.60 ,
random_state =0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis')
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

kmeans = KMeans(n_clusters=3): Здесь создается объект kmeans класса KMeans. Ему передается аргумент n_clusters=3, который указывает, что алгоритм K-средних должен разделить данные на три кластера.
kmeans.fit(X): Этот вызов метода fit выполняет обучение модели K-средних на ваших данных X, что означает, что алгоритм будет пытаться найти оптимальные центры кластеров.

```python
# Classification of data (three clusters)
kmeans = KMeans(n_clusters =3, n_init=10)
kmeans.fit(X)
```

centers = kmeans.cluster_centers_: После выполнения обучения модели, этот код извлекает координаты центров кластеров, найденных алгоритмом K-средних, и сохраняет их в переменной centers. Эти центры представляют собой точки в пространстве признаков, около которых сгруппированы данные в каждом кластере.

```python
# Determining the centroids
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(centers[:, 0], centers[:,1], c={'green', 'yellow', 'blue'}, s=100, alpha=0.7)
plt.title("Исходные данные с истинными метками")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```
